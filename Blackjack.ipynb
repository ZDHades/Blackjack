### Rules:

`1. ` The game will have two players: the Dealer and the Player. The game will start off with a deck of 52 cards. The 52 cards will consist of 4 different suits: Clubs, Diamonds, Hearts and Spades. For each suit, there will be cards numbered 1 through 13. <br>
**Note: No wildcards will be used in the program**

`2. ` When the game begins, the dealer will shuffle the deck of cards, making them randomized. After the dealer shuffles, it will deal the player 2 cards and will deal itself 2 cards from. The Player should be able to see both of their own cards, but should only be able to see one of the Dealer's cards.
 
`3. ` The objective of the game is for the Player to count their cards after they're dealt. If they're not satisfied with the number, they have the ability to 'Hit'. A hit allows the dealer to deal the Player one additional card. The Player can hit as many times as they'd like as long as they don't 'Bust'. A bust is when the Player is dealt cards that total more than 21.

`4. ` If the dealer deals the Player cards equal to 21 on the **first** deal, the Player wins. This is referred to as Blackjack. Blackjack is **NOT** the same as getting cards that equal up to 21 after the first deal. Blackjack can only be attained on the first deal.

`5. ` The Player will never see the Dealer's hand until the Player chooses to 'stand'. A Stand is when the player tells the dealer to not deal it anymore cards. Once the player chooses to Stand, the Player and the Dealer will compare their hands. Whoever has the higher number wins. Keep in mind that the Dealer can also bust. 

```python
import random
from IPython.display import clear_output
class Deck:
    def __init__(self, cards=[]):
        self.cards = [(rank,suit) for suit in '♠ ♢ ♣ ♡'.split() for rank in [str(n) for n in range(2, 11)] + list('JQKA')]
    def deal(self):
        return self.cards.pop(random.randrange(len(self.cards)))
        
class Player:
    def __init__(self, hand = None, score = 0):
        self.hand = []
        self.score = score
    def Hit(self, deck):
        self.hand.append(deck.deal())
        for _ in self.hand:
            if _[0] == "K" or "Q" or "J":
                self.score += 10
            if _[0] == "A":
                self.score += 1
            else:
                self.score += int(_[0])        

class Dealer(Player):
    def DD(self, deck):
        while self.score < 17:
            self.hand.append(deck.deal)
            return self.score

    def show_dealer():
        return self.hand[0]
    

class Human(Player):
    pass
  
        
class Game:
    def __init__(self, Human, dealer, Deck):
        self.Human = Human
        self.Dealer = dealer
        self.Deck = Deck
        
    def initial_deal(self):
        self.Human.Hit(self.Deck)
        self.Human.Hit(self.Deck)
        Dealer.DD(self.Deck)
        print(f"Your cards are {self.Human.cards} that add up to a value of {self.Human.score}")
        print(f"The dealers frist card is {self.Dealer.hand[0]}")

def main():
    flag1 = False
    print("Welcome to Hades' Casino Games! Today's feature game is Blackjack!")
    while not flag1:
        deck = Deck()
        player1 = Human()
        dealer1 = Dealer()
        game = Game(player1, dealer1, deck)
        game.initial_deal()
        flag2 = False
        while not flag2:
            if player1.score == 21:
                print("You got 21! Congratz")
                flag2 = True
            else:
                flag3 = False
                while flag3:
                    HS = input("Would you like to hit or stand?").lower()
                    if HS == "hit":
                        player1.Hit(deck)
                        print(f"your new hand is {player1.hand}")
                        print(f"your new total is {player1.score}")
                        if player1.score > 21:
                            flag3 = True
                        if player1.score == 21:
                            flag3 = True
                            flag2 = True
                        else:
                            continue
                    if HS == "stand":
                        flag3 = True
                        flag2 = True
                    else:
                        print("Invalid choice, choose hit or stand")
        clear_output()
        print(f"You had a final score of {player1.score}")
        print(f"The dealer had a final score of {dealer1.score}")
        results = False
        while not results:
            if dealer1.score and player1.score == 21:
                print("You and the dealer both got 21! Its a push!")
                results = True
            elif player1.score > 21:
                print(f"Your score was {player1.score}, you busted! Dealer wins!")
                results = True
            elif player1.score == 21:
                print("You got Blackjack! Congratz!")
                results = True
            elif dealer1.score > player1.score:
                print(f"You got {player1.score} and the dealer got {dealer1.score}. Dealer wins!")
                results = True
            elif dealer1.score < player1.score:
                print(f"You got {player1.score} and the dealer got {dealer1.score}. You win!")
                results = True
        retry = False
        while not retry:
            _ = input("Would you like to play again? y/n")
            if _ == "y":
                retry = True
            if _ == "n":
                retry = True
                flag1 = True
            else:
                print("Enter y/n please")
            
```
